```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import data
body_data = pd.read_csv('weights_and_heights.csv')
body_data.head()
```

```python
# inspect data
print(body_data.describe())
print('\n')
print(body_data['Gender'].value_counts())
```

```python
# split data
male = body_data[body_data['Gender'] == 'Male']
female = body_data[body_data['Gender'] == 'Female']
```

```python
# plot male and female weight histogram
plt.style.use('seaborn-poster')

plt.title('Distribution of Body Weight by Gender')
plt.xlabel('Weight (lbs)')
plt.ylabel('Frequency')


plt.hist(
    male['Weight'], 
    bins = 20, 
    edgecolor = 'black', 
    alpha = 0.6,
    color = 'royalblue',
    label = 'Male'
)

plt.hist(
    female['Weight'], 
    bins = 20, 
    edgecolor = 'black', 
    alpha = 0.6,
    color = 'magenta',
    label = 'Female'
)

plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
# create sample data
x_values = list(range(1, 11))
y_values = [i ** 2 for i in x_values]
z_values = [i ** 3 for i in x_values]

print(x_values)
print(y_values)
print(z_values)
```

```python
# create plot and style (for demonstration purposes)
plt.plot(
    x_values, y_values, 
    label = 'X Squared', 
    color = 'deeppink', # search matplotlib colors
    linewidth = 2, # default 1
    linestyle = '--' # default '-'
)

plt.plot(
    x_values, z_values, 
    label = 'X Cubed', 
    color = '#0000FF', # hex code
    linewidth = 2,
    linestyle = ':',
    marker = 'o', markersize = 10, markerfacecolor = 'red', markeredgecolor = 'green'
)

plt.title("Exponential Growth")
plt.xlabel('The Values of X')
plt.ylabel('The Values of Y')

plt.legend()
plt.show()
```

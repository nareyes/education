```python
import pandas as pd
import numpy as np
```

```python
customer_details = pd.read_excel('grocery_database.xlsx', sheet_name = 'customer_details')
customer_details.head()
```

```python
# sort ascending
customer_details.sort_values(
    by = 'distance_from_store', 
    inplace = True, 
    ascending = True
)

customer_details.head()
```

```python
# sort descending
customer_details.sort_values(
    by = 'distance_from_store', 
    inplace = True, 
    ascending = False
)

customer_details.head()
```

```python
# sort multiple columns
customer_details.sort_values(
    by = ['distance_from_store', 'credit_score'], 
    inplace = True, 
    ascending = True
)

customer_details.head()
```

```python
# sort na first
customer_details.sort_values(
    by = ['distance_from_store', 'credit_score'], 
    inplace = True, 
    ascending = True, 
    na_position = 'first'
)

customer_details.head()
```

```python
# ranking (create data frame)
df = pd.DataFrame({
    'col1': [1, 1, 1, 2, 3, 4, 5, np.nan, 6, 8]
})

df
```

```python
# ranking
df['col1_rank'] = df['col1'].rank()
df
```

```python
# ranking methods
df['col1_avg_rank'] = df['col1'].rank(method = 'average') # default, applies average to ties
df['col1_min_rank'] = df['col1'].rank(method = 'min') # applies min rank for ties
df['col1_max_rank'] = df['col1'].rank(method = 'max') # applies max rank for ties
df['col1_first_rank'] = df['col1'].rank(method = 'first') # applies priority by order
df['col1_dense_rank'] = df['col1'].rank(method = 'dense') # ties preserved, rank continous

df
```

```python
# ranking and nan values
df['col1_dense_rank_na_top'] = df['col1'].rank(method = 'dense', na_option = 'top') # ranks nan first
df['col1_dense_rank_na_bottom'] = df['col1'].rank(method = 'dense', na_option = 'bottom') # ranks nan last

df
```

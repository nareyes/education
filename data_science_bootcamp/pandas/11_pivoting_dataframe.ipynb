```python
import pandas as pd
```

```python
# import data
transactions = pd.read_excel('grocery_database.xlsx', sheet_name = 'transactions')
transactions.head()
```

```python
# import data
product_areas = pd.read_excel('grocery_database.xlsx', sheet_name = 'product_areas')
product_areas.head()
```

```python
# merge data
transactions = pd.merge(transactions, product_areas, how = 'inner', on = 'product_area_id')
transactions.head()
```

```python
# previous example using groupby
sales_summary = transactions.groupby(['transaction_date', 'product_area_name'])['sales_cost'].sum().reset_index()
sales_summary
```

```python
# pivoting data and applying an aggregate function
sales_summary_pivot = transactions.pivot_table(
    index = 'transaction_date', 
    columns = 'product_area_name', 
    values = 'sales_cost', 
    aggfunc = 'sum',
    fill_value = 0,
    margins = True,
    margins_name = 'Total'
)
sales_summary_pivot
```

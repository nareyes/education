```python
import pandas as pd
```

```python
# create data frames
df_a = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})

df_b = pd.DataFrame({
    'C': [1, 2, 3],
    'D': [4, 5, 6]
})

# concatenating two data frames on columns
df_c = pd.concat([df_a, df_b], axis = 1)
df_c
```

```python
# create data frames (same column names)
df_a = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})

df_b = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})

# concatenating two data frames on index (can also append but concat has better functionality)
df_c = pd.concat([df_a, df_b], axis = 0)
df_c
```

```python
# create data frames for merging (sql joins)
df_a = pd.DataFrame({
    'user_id': [1, 2, 3, 5, 7],
    'age': [41, 15, 60, 18, 29]
})

df_b = pd.DataFrame({
    'user_id': [1, 2, 3, 4, 5],
    'gender': ['M', 'F', 'F', 'F', 'M']
})
```

```python
#  merging data frames with inner join
df_inner = pd.merge(df_a, df_b, how = 'inner', on = 'user_id')
df_inner
```

```python
#  merging data frames with left join
df_left = pd.merge(df_a, df_b, how = 'left', on = 'user_id')
df_left
```

```python
#  merging data frames with right join
df_right = pd.merge(df_a, df_b, how = 'left', on = 'user_id')
df_right
```

```python
#  merging data frames with outer join
df_outer = pd.merge(df_a, df_b, how = 'outer', on = 'user_id')
df_outer
```

### **Additional Notes**

#### - *[Join Diagrams](https://www.skytowner.com/explore/pandas_dataframe_join_method)*
#### - *To join on multiple column names, pass in a list to the 'on' parameter*
#### - *To join on different column names, use the 'left_on' and 'right_on' parameters*
#### - *The default join is inner join, and Pandas will scan data frames for a shared column if one is not supplied. This is not a best practice!*

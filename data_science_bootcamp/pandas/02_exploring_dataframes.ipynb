```python
import pandas as pd
```

```python
# import transactions data
transactions = pd.read_excel('grocery_database.xlsx', sheet_name = 'transactions')
transactions
```

```python
# view shape (rows, columns)
transactions.shape #attributes don't require parenthesis 
```

```python
# view head (top n rows)
transactions.head(n = 3) # default n = 5
```

```python
# view tail (bottom n rows)
transactions.tail(n = 3) # default n = 5
```

```python
# view sample (random)
transactions.sample(n = 3)
transactions.sample(frac = 0.1) # 10% of data frame
```

```python
# descriptive statistics
transactions.describe()
```

```python
# view largest values
transactions.nlargest(5, 'sales_cost')
```

```python
# view smallest values
transactions.nsmallest(5, 'sales_cost')
```

```python
# view distinct counts
transactions.nunique()
```

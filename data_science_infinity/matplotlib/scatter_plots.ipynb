```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import data
body_data = pd.read_csv('weights_and_heights.csv')
body_data.head()
```

```python
# inspect data
print(body_data.describe())
print('\n')
print(body_data['Gender'].value_counts())
```

```python
# split data
male = body_data[body_data['Gender'] == 'Male']
female = body_data[body_data['Gender'] == 'Female']
```

```python
# sample data
male_sample = male.sample(200)
female_sample = female.sample(200)
```

```python
# plot male and female weight v height scatter plot
plt.style.use('seaborn-poster')

plt.title('Male Weight vs. Height')
plt.xlabel('Weight (lbs)')
plt.ylabel('Height (in)')

plt.scatter(
    male_sample['Weight'], male_sample['Height'], # x, y
    color = 'blue',
    s = 700, # size
    alpha = 0.5
)

plt.axvline(x = male_sample['Weight'].median(), color = 'black', linestyle = '--')
plt.axhline(y = male_sample['Height'].median(), color = 'black', linestyle = '--')
plt.tight_layout()
plt.show()
```

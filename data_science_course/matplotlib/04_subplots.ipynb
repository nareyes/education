```python
import matplotlib.pyplot as plt
```

```python
# create sample data
x_values = list(range(1, 11))
y_values = [i ** 2 for i in x_values]
z_values = [i ** 3 for i in x_values]

print(x_values)
print(y_values)
print(z_values)
```

```python
# create vertical subplots and style with preset matplotlib styles
plt.style.use('seaborn-notebook')

plt.subplot(2, 1, 1)
plt.plot(x_values, y_values, label = 'X Squared')
plt.title("Squared Values")
plt.xlabel('The Values of X')
plt.ylabel('The Values of X Squared')

plt.subplot(2, 1, 2)
plt.bar(x_values, z_values, label = 'X Cubed')
plt.title("Cubed Values")
plt.xlabel('The Values of X')
plt.ylabel('The Values of X Cubed')

plt.tight_layout()
plt.show()
```

```python
# create four subplots and style with preset matplotlib styles
plt.style.use('seaborn-notebook')

plt.subplot(2, 2, 1)
plt.plot(x_values, y_values, label = 'X Squared')
plt.title("Squared Values")
plt.xlabel('The Values of X')
plt.ylabel('The Values of X Squared')

plt.subplot(2, 2, 2)
plt.bar(x_values, z_values, label = 'X Cubed')
plt.title("Cubed Values")
plt.xlabel('The Values of X')
plt.ylabel('The Values of X Cubed')

plt.subplot(2, 2, 3)
plt.plot(x_values, y_values, label = 'X Squared')
plt.title("Squared Values")
plt.xlabel('The Values of X')
plt.ylabel('The Values of X Squared')

plt.subplot(2, 2, 4)
plt.bar(x_values, z_values, label = 'X Cubed')
plt.title("Cubed Values")
plt.xlabel('The Values of X')
plt.ylabel('The Values of X Cubed')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
# create sample data
x_values = list(range(1, 11))
y_values = [i ** 2 for i in x_values]
z_values = [i ** 3 for i in x_values]

print(x_values)
print(y_values)
print(z_values)
```

```python
# create plot and format
plt.plot(x_values, y_values, label = 'X Squared')
plt.plot(x_values, z_values, label = 'X Cubed')

plt.title("Exponential Growth")

plt.xlabel('The Values of X')
plt.xlim(0, 10)
plt.xticks(range(11), rotation = 45)

plt.ylabel('The Values of Y')
plt.ylim(0, 1000)

plt.grid(True)
plt.legend()
plt.show()
```

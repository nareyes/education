```python
import pandas as pd
```

```python
# import data
customer_details = pd.read_excel('grocery_database.xlsx', sheet_name ='customer_details')
customer_details.head()
```

```python
# import data
product_areas = pd.read_excel('grocery_database.xlsx', sheet_name ='product_areas')
product_areas.head()
```

```python
# map new column from existing column
# pandas will use nan for values that are not mapped
customer_details['gender_numeric'] = customer_details['gender'].map({ # dict old_val: new_val
    'M': 0,
    'F': 1
})
customer_details.head()
```

```python
# map new column from existing column using replace
# pandas will use the existing value for values that are not mapped
customer_details['gender_numeric'] = customer_details['gender'].replace({ # dict old_val: new_val
    'F': 1
})
customer_details.head()
```

```python
# apply a function to every value in a series or data frame
product_areas['product_area_name_length'] = product_areas['product_area_name'].apply(len)
product_areas.head()
```

```python
# apply a created function to every value in a series or data frame
def update_profit_margin(profit_margin):
    if profit_margin > 0.2:
        return profit_margin * 1.2
    else:
        return profit_margin * 0.8

product_areas['profit_margin_updated'] = product_areas['profit_margin'].apply(update_profit_margin)
product_areas.head()
```

```python
# apply a function to every element in a data frame using applymap
new_df = pd.DataFrame({
    'A': [1, 2],
    'B': [3, 4],
    'C': [5, 6]
})

def square(n):
    return n ** 2

new_df.applymap(square)
```

```python
import pandas as pd
import numpy as np
```

```python
# import data
transactions = pd.read_excel('grocery_database.xlsx', sheet_name = 'transactions')
transactions.head()
```

```python
# adding column with a static value
transactions['store_id'] = 1
transactions.head()
```

```python
# adding column with a dynamic value
transactions['target_profit'] = transactions['sales_cost'] * 0.2
transactions.head()
```

```python
# adding column with conditional logic
transactions['sales_type'] = np.where(transactions['sales_cost'] > 20, 'Large', 'Small')
transactions.head()
```

```python
# adding column with conditional logic (using lists)
condition_rules = [
    transactions['sales_cost'] > 50,
    transactions['sales_cost'] > 20,
    transactions['sales_cost'] > 10
]

outcomes = [
    'X-Large',
    'Large',
    'Medium'
]

transactions['sales_type'] = np.select(condition_rules, outcomes, default = 'Small')
transactions.head()
```

```python
# removing columns
new_df = transactions.drop(['store_id'], axis = 1) # axis = 1 specifies columns
new_df.head()
```

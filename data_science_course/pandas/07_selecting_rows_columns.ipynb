```python
import pandas as pd
```

```python
transactions = pd.read_excel('grocery_database.xlsx', sheet_name = 'transactions')
transactions.head()
```

```python
# return all column lables
list(transactions)
```

```python
# iloc rows (index)
# df.iloc[row_indexes]
transactions.iloc[0]
transactions.iloc[0:4]
transactions.iloc[[0, 1, 2, 3, 4]]
```

```python
# iloc rows and columns (index)
# df.iloc[row_indexes, column_indexes]
transactions.iloc[:, :] # empty split will return all rows/columns
transactions.iloc[0:5, [0, 2, -1]]
```

```python
# loc rows (labels)
# df.loc[row_labels, column_labels]
transactions.loc[0:4, ['customer_id', 'product_area_id', 'sales_cost']]
```

```python
# conditional logic (single)
transactions.loc[transactions['customer_id'] == 642, ['customer_id', 'sales_cost']].head()
```

```python
# conditional logic (multiple)
transactions.loc[
    (transactions['customer_id'] == 646) & # pipe operator for or condition
    (transactions['num_items'] > 5)
].head()
```

```python
# conditional logic (isin)
transactions.loc[transactions['customer_id'].isin([642, 700])].head()
```

```python
# conditional logic (not in)
transactions.loc[~transactions['customer_id'].isin([642, 700])].head() # tilda indicates opposite
```

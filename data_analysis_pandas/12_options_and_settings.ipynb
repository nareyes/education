```python
import pandas as pd
import numpy as np
```

# Generate Fake Data Set

```python
# generate data with numpy
data = np.random.randint(
    low = 0,
    high = 100,
    size = [1000, 50]
)

df = pd.DataFrame(data)
df.head()
```

Options w/ Attributes & Dot Syntax

```python
# set max row option
pd.options.display.max_rows = 10

# set max column option
pd.options.display.max_columns = 10

# display df with new settings
df
```

# Options w/ Methods

```python
# set max row option
pd.set_option('display.max_rows', 20)

# set max column option
pd.set_option('display.max_columns', 20)

# display df with new settings
df
```

# Precision Option

```python
# create data
df = pd.DataFrame(np.random.randn(5, 5))
df
```

```python
# get precision option
pd.get_option('display.precision')
```

```python
# set precision option
pd.set_option('display.precision', 2)

# display df with new settings
df
```

```python
# resetting precision option
pd.reset_option('display.precision')

# display df with new settings
df
```

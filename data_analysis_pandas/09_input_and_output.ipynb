```python
import pandas as pd
import xlrd
import openpyxl
```

# URL

```python
# passing a url to read_csv method
url = 'https://data.cityofnewyork.us/api/views/25th-nujf/rows.csv'
baby_names = pd.read_csv(url)

baby_names.head()
```

# Object Conversions

```python
# convert series to data frame
baby_names["Child's First Name"].to_frame().head()
```

```python
# convert series to list
baby_names["Child's First Name"].tolist()
```

```python
# convert series to dictionary
# dictionaries do not allow duplicate keys, so make sure index is unique
baby_names["Child's First Name"].to_dict()
```

```python
# convert series to string
# pass comma and space as a delimeter
# sort, remove duplicates, convert to title
', '.join(baby_names["Child's First Name"].sort_values().drop_duplicates().str.title())
```

# Export to CSV

```python
# export data frame to csv
baby_names.to_csv('data/baby_names.csv', index = True) # keep index (default)
baby_names.to_csv('data/baby_names_no_index.csv', index = False) # remove index
```

```python
# export data frame with specified columns to csv
baby_names.to_csv(
    'data/baby_names_filtered.csv',
    columns = ['Gender', 'Ethnicity', "Child's First Name"],
    index = False) # remove index
    # encoding = 'utf-8' if you recieve an encoding error
```

# Exporting Excel Files

```python
# split baby names data frame by gender
male_baby_names = baby_names[baby_names['Gender'] == 'MALE']
female_baby_names = baby_names[baby_names['Gender'] == 'FEMALE']

# baby_names[male_baby_names]
# baby_names[female_baby_names]
```

```python
# instantiate excel writer
excel_file = pd.ExcelWriter('data/baby_names_by_gender.xlsx')
```

```python
# write to excel file
male_baby_names.to_excel(excel_file, sheet_name = 'Male', index = False)
female_baby_names.to_excel(excel_file, sheet_name = 'Female', index = False)
excel_file.save()
```
